## Create file map the Acession Number to _Sham Number

```
with open("/content/old_elvos.key.csv", "r") as csv_file:
  csv_reader = csv.DictReader(csv_file)

  with open("Acess_to_Sham.csv", "w") as new_file:
    fieldnames = ["_ShamID", "AccessionNumber"]

    csv_writer = csv.DictWriter(new_file, fieldnames=fieldnames, delimiter=",")

    csv_writer.writeheader()

    for line in csv_reader:
      for item in elvo_list:
        del line[item]
      csv_writer.writerow(line)
```
