## CSV dictionary Writer

```
with open("/content/old_elvos.key.csv", "r") as csv_file:
  csv_reader = csv.DictReader(csv_file)

  count = 0

  for line in csv_reader:
    if count < 10:
      print(line)
      print(line["_ShamID"])
      print(line["AccessionNumber"])
      count += 1
      elvlo_keys = line.keys()
      print(elvlo_keys)
      print("\n ")
```

```
elvo_list = list(elvlo_keys)
print(elvo_list)
type(elvo_list)
```

```
elvo_list.remove("_ShamID")
elvo_list.remove("AccessionNumber")
```

```
print(elvo_list)
```

## Create file map the Acession Number to _Sham Number

```
with open("/content/old_elvos.key.csv", "r") as csv_file:
  csv_reader = csv.DictReader(csv_file)

  with open("Acess_to_Sham.csv", "w") as new_file:
    fieldnames = ["_ShamID", "AccessionNumber"]

    csv_writer = csv.DictWriter(new_file, fieldnames=fieldnames, delimiter=",")

    csv_writer.writeheader()

    for line in csv_reader:
      for item in elvo_list:
        del line[item]
      csv_writer.writerow(line)
```
